#### Question : 
##### Write a python program that takes an input number from the user, Converts it to different numeric data types(integers,float, and complex), and display the converted values.


```python
# Get user input for a number
num = input("Enter a number: ")

# Convert the input to integer, float, and complex data types
try:
    # convert the input to an integer
    # int function converts a number to an integer, which is a whole number without a fractional part
    int_num = int(num)
    print(f"Integer: {int_num}")
    
    #convert the input to a float
    # float function converts a number to a floating-point number, which is a number with a fractional part
    float_num = float(num)
    print(f"Float: {float_num}")
    
    # convert the input to a complex number
    # complex function converts a number to a complex number, which is a number with real and imaginary parts
    complex_num = complex(num)
    print(f"Complex: {complex_num}")
except ValueError:
    print("Invalid input. Please enter a valid number.")
```

#### Question : 
##### Write a Python program that asks the user for a number and determines whether it is positive, negative, or zero.

```python
while True:
    # Ask the user for a number or 'exit' to quit
    user_input = input("Enter a number (or type 'exit' to quit): ")

    # Check  the user wants to exit
    if user_input.lower() == 'exit':
        print("Exiting the program.")
        break

    # Validate  the input is a number
    try:
        number = float(user_input)
    except ValueError:
        print("Invalid input! Please enter a valid number or 'exit' to quit.")
        continue

    # Determine  the number is positive, negative, or zero
    if number > 0:
        print("The number is positive.")
    elif number < 0:
        print("The number is negative.")
    else:
        print("The number is zero.")
```

#### Question : 
##### Create a Python script that takes two numbers as input and prints whether both numbers are even, odd, or one of each using relational and logical operators.

```python
# check  number is even
def is_even(num):
    return num % 2 == 0

# check  number is odd
def is_odd(num):
    return num % 2 != 0

# Input two numbers 
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Check the parity of both numbers
if is_even(num1) and is_even(num2):
    print("Both numbers are even.")
elif is_odd(num1) and is_odd(num2):
    print("Both numbers are odd.")
else:
    print("One number is even and the other is odd.")
```

#### Question : 
##### Write a Python program that takes an integer input and prints its binary, octal, and hexadecimal equivalents using a for loop and bitwise operators.


```python
# convert integer to binary
def to_binary(n):
    if n == 0:
        return '0'
    binary_str = ''
    for _ in range(n.bit_length()): 
        binary_str = str(n & 1) + binary_str
        n >>= 1
    return binary_str

# convert integer to octal 
def to_octal(n):
    if n == 0:
        return '0'
    octal_str = ''
    for _ in range((n.bit_length() + 2) // 3):  
        octal_str = str(n & 7) + octal_str
        n >>= 3
    return octal_str

# convert integer to hexadecimal
def to_hexadecimal(n):
    hex_chars = '0123456789ABCDEF'
    if n == 0:
        return '0'
    hexadecimal_str = ''
    for _ in range((n.bit_length() + 3) // 4): 
        hexadecimal_str = hex_chars[n & 15] + hexadecimal_str
        n >>= 4
    return hexadecimal_str

# Input an integer 
num = int(input("Enter an integer: "))

# Convert and print the number in binary, octal, and hexadecimal
binary_rep = to_binary(num)
octal_rep = to_octal(num)
hexadecimal_rep = to_hexadecimal(num)

print(f"Binary: {binary_rep}")
print(f"Octal: {octal_rep}")
print(f"Hexadecimal: {hexadecimal_rep}")
```
